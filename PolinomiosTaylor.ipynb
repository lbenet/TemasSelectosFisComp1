```julia
immutable Taylor{T}
    coeffs :: Array{T,1}
    order :: Int
    # Constructor interno
    function Taylor(coeffs::Array{T,1}, order::Int)
        ll = length(coeffs)
        order = max(ll-1, order)
        v = zeros(T,order+1)
        v[1:ll] = coeffs[1:ll]
#         for i=1:ll
#             v[i] = coeffs[i]
#         end
        new(v, order)
    end
end
```

```julia
# Constructores externos
Taylor{T<:Number}(v::Array{T,1}, n::Int) = Taylor{T}(v,n)
Taylor{T<:Number}(v::Array{T,1}) = Taylor{T}(v,0)
Taylor{T<:Number}(c::T, n::Int) = Taylor{T}([c],n)
Taylor{T<:Number}(c::T) = Taylor{T}([c],0)
Taylor{T<:Number}(t::Taylor{T},n) = Taylor{T}(t.coeffs, n)
Taylor{T<:Number}(t::Taylor{T}) = Taylor{T}(t.coeffs, 0)
```

```julia
function +(f::Taylor,g::Taylor)
    l = max(f.order,g.order)
    f, g = Taylor(f,l), Taylor(g,l)
    Taylor(f.coeffs+g.coeffs, l)
end
+(f::Taylor) = f
function -(f::Taylor,g::Taylor)
    l = max(f.order,g.order)
    f, g = Taylor(f,l), Taylor(g,l)
    Taylor(f.coeffs-g.coeffs, l)
end
-(f::Taylor) = Taylor(-f.coeffs,f.order)
```

```julia
function *(f::Taylor,g::Taylor)
    ll = max(f.order,g.order)
    f, g = Taylor(f,ll), Taylor(g,ll)
    T = promote_type( eltype(f.coeffs), eltype(g.coeffs) )
    v = zeros(T,ll+1)
    for ord = 0:ll  ## orden del resultado
        suma = zero(T)
        for n = 0:ord  ## implementación \Sum
            suma += f.coeffs[n+1] * g.coeffs[ord-n+1]
        end
        v[ord+1] = suma
    end
    #
    Taylor(v, ll)
end
```

```julia
+(t::Taylor,c::Number) = t + Taylor(c)
-(t::Taylor,c::Number) = t - Taylor(c)
+(c::Number,t::Taylor) = Taylor(c) + t
-(c::Number,t::Taylor) = Taylor(c) - t

*(t::Taylor,c::Number) = Taylor(c*t.coeffs, t.order)
*(c::Number,t::Taylor) = Taylor(c*t.coeffs, t.order)

/(t::Taylor,c::Number) = inv(c)*t
```

## Integración por Taylor

\begin{equation}
\dot{x} = - x^2, x(0) = 1
\end{equation}

```julia
ff(x) = -x*x
```

```julia
eps(1.0)
```

```julia
const ordenMax=18
const epsilon = 1e-20

c0 = 1.0
tiempo = 0.0
```

```julia
function nexttaylor(c0::Real)
    v = [c0]
    for ord = 1: ordenMax
        q0 = Taylor(v)
        c1 = ff(q0).coeffs[ord]/ord
        push!(v,c1)
    end
    v
end
```

```julia
nexttaylor(1.0)
```

```julia
function timestep(v)
    pot = inv(ordenMax)
    h1 = (epsilon/abs(v[end]))^pot
    pot = inv(ordenMax-1)
    h2 = (epsilon/abs(v[end-1]))^pot
    min(h1,h2)
end
```

```julia
h = timestep(v)
```

```julia
function evalhorner(v,h)
    ll = length(v)
    suma = v[end]
    for i = ll-1:-1:1
        suma = v[i] + h*suma
    end
    suma
end
```

```julia
evalhorner(v,h)
```

```julia
function taylorStepper(c0,tiempo)
    v = nexttaylor(c0)
    h = timestep(v)
    c1 = evalhorner(v,h)
    tiempo += h
    return c1, tiempo
end
```

```julia
x0 = 1.0
tiempo = 0.0
while tiempo < 10.0
    x0, tiempo = taylorStepper(x0,tiempo)
    println(tiempo ,"\t", x0, "\t", abs(inv(1.0+tiempo)-x0)/eps(1.0))
end
```

```julia
function nexttaylor1(c0::Real,ff::Function)
    v = [c0]
    for ord = 1: ordenMax
        q0 = Taylor(v)
        c1 = ff(q0).coeffs[ord]/ord
        push!(v,c1)
    end
    v
end

function taylorStepper1(c0,tiempo)
    v = nexttaylor1(c0,ff)
    h = timestep(v)
    c1 = evalhorner(v,h)
    tiempo += h
    return c1, tiempo
end
```

```julia
ff(x) = x*x
x0 = 1.0
tiempo = 0.0
for i=1:100
    x0, tiempo = taylorStepper1(x0,tiempo)
    println(tiempo ,"\t", x0, "\t")
end
```
