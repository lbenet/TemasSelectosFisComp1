# Empleo de la exponencial de Liapunov para la bifurcación de sistemas no lineales   Enoc Guerrero Garay

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def logistic(r, x):
    return r*x*(1-x)
```

```python
n = 15000
r = np.linspace(2.5, 4.0, n)
```

# Ahora simularemos 15000 iterationes del mapa logistico, y mantendremos las ultimas 100 para el diagrama de bifurcacion 

```python
interactiones = 1500
last = 100
```

```python
x = 1e-5 * np.ones(n)
```

```python
lyapunov = np.zeros(n)
```

```python
plt.figure(figsize=(6,7));
plt.subplot(211);
for i in range(iterations):
    x = logistic(r, x)
    # La suma parcial del exponencial de Lyapunov
    lyapunov += np.log(abs(r-2*r*x))
    # Mostraremos el Diagrama de Bifurcación
    if i >= (iterations - last):
        plt.plot(r, x, ',k', alpha=.04)
plt.xlim(2.5, 4);
plt.title("Diagrama de Bifurcacion");

# Mostraremos la exponencial de Lyapunov.
plt.subplot(212);
plt.plot(r[lyapunov<0], lyapunov[lyapunov<0] / iterations,
         ',k', alpha=.2);
plt.plot(r[lyapunov>=0], lyapunov[lyapunov>=0] / iterations,
         ',r', alpha=.5);
plt.xlim(2.5, 4);
plt.ylim(-2, 1);
plt.title("Exponencial de Lyapunov");
plt.tight_layout();
```
