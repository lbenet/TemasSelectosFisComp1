# Redes booleanas

Una red booleana es un conjunto de variables booleanas conectadas entre sí mediante funciones lógicas (and or not). Este tipo de redes pueden clasificarse según su manera de evolucionar, me centraré en las redes booleanas con evolución sincrónica.

Que una red booleana tenga evolución sincrónica quiere decir que todas sus variables se actualizan al mismo tiempo y que estas dependen únicamente del estado anterior de la red. Esto impone una dinámica discreta a la red puesto que para actualizarse, las variables tendrán que referirse al estado inmediato anterior, es decir, si $\{ \sigma_i \}_{i\in \{ 1,2,...,n \} }$ son las $n$ variables que conforman la red, y denotamos como $\sigma^t _i$ a la variable $\sigma_i$ al tiempo t, tenemos que
$$\sigma^{t+1} _i=\sigma^{t+1} _i (\sigma^{t} _1,\sigma^{t} _2,...,\sigma^{t} _i,...,\sigma^{t} _n)$$
Definamos una red de n variables, conectividad m (es decir, cada variable en un tiempo $t+1$ es función de otras m a tiempo $t$), una matriz de conectividad que nos dice con qué variables se relaciona cada una y otra que nos indica la función lógica que las relaciona.


```julia
n=100
var=rand(0:1,n)
m=2
conect=rand((1:n),n,m)
yo=rand(0:5,n,m-1)
using PyPlot
```

Ahora definamos la función not, una función para actualizar el estado de la red en un "paso" y otra, basada en la anterior, para actualizarla en todos los pasos deseados.

```julia
function not(a::Integer)
    if a==0
        a=1
    else
        a=0
    end
end

function paso2(var::Array, conect::Array, yo::Array)
    nvar=zeros(Int64,length(var))
    for i=1:length(var)
        if yo[i]==0
            nvar[i]=var[conect[i,1]] | var[conect[i,2]]
        end
        if yo[i]==1
            nvar[i]=var[conect[i,1]] & var[conect[i,2]]
        end
        if yo[i]==2
            nvar[i]=not(var[conect[i,1]]) & var[conect[i,2]]
        end
        if yo[i]==3
            nvar[i]=var[conect[i,1]] & not(var[conect[i,2]])
        end
        if yo[i]==4
            nvar[i]=not(var[conect[i,1]]) | var[conect[i,2]]
        end
        if yo[i]==5
            nvar[i]=var[conect[i,1]] | not(var[conect[i,2]])
        end
    end
    return nvar
end

function pasos2(var::Array, conect::Array, yo::Array, n::Integer)
    for j=1:n
        var=pasob(var,conect,yo)
    end
    return var
end
```

Notemos que lo anterior sólo funciona para el caso m=2, hagamos otra función para el caso general.

```julia
function paso(var::Array, conect::Array, yo::Array)
    nvar=zeros(Int64,length(var),size(yo,2))
    nvar[1:length(var)]=paso2(var,conect,yo)
    for i=1:length(var)
        for j=2:size(yo,2)
            if yo[i,j]==0
                nvar[i,j]=nvar[i,j-1] | var[conect[i,j+1]]
            end
            if yo[i,j]==1
                nvar[i,j]=nvar[i,j-1] & var[conect[i,j+1]]
            end
            if yo[i,j]==2
                nvar[i,j]=not(nvar[i,j-1]) & var[conect[i,j+1]]
            end
            if yo[i,j]==3
                nvar[i,j]=nvar[i,j-1] & not(var[conect[i,j+1]])
            end
            if yo[i,j]==4
                nvar[i,j]=not(nvar[i,j-1]) | var[conect[i,j+1]]
            end
            if yo[i,j]==5
                nvar[i,j]=nvar[i,j-1] | not(var[conect[i,j+1]])
            end
        end
    end
    return nvar[1:length(var),size(yo,2)]
end

function pasos(var::Array, conect::Array, yo::Array, n::Integer)
    for j=1:n
        var=paso(var,conect,yo)
    end
    return var
end
```

Investiguemos cómo cambia el número de variables prendidas con los pasos y su relación con la conectividad y con el número de variables

```julia
function onoff(var::Array, conect::Array, yo::Array, n::Integer)
    a=zeros(n+1,2)
    for i=0:n
        a[i+1,1]=i
        b=pasos(var,conect,yo,i)
        for j=1:length(var)
           a[i+1,2]=a[i+1,2]+b[j] 
        end
        a[i+1,2]=(a[i+1,2]/length(var))*100
    end
    return a
end
```

```julia
n=100
var=rand(0:1,n)
m=2
conect=rand((1:n),n,m)
yo=rand(0:5,n,m-1)
a=onoff(var,conect,yo,100)
plot(a[1:60,1],a[1:60,2])
title("n=100, m=2")
ylabel("Porcentaje de variables activadas")
xlabel("Pasos")
```

Obteneruna regla de predicción (aunque sea cualitativa) del sistema en función n y m es complicado pues el sistema puede ser cíclico, constante, o ninguna de las anteriores.

Hablemos ahora de la robustez que tienen estos sistemas y su relación con n y m.

```julia
function onoffp(var::Array, conect::Array, yo::Array, n::Integer, h::Integer)
    #h es el número de variables en cada conjunto
    p=int32(length(var)/h)
    a=zeros(n+1,p+1)
    for i=0:n
        a[i+1,1]=i
        b=pasos(var,conect,yo,i)
        for k=1:p
            for j=1+((k-1)*h):h+((k-1)*h)
                a[i+1,k+1]=a[i+1,k+1]+b[j]
            end
            a[i+1,k+1]=(a[i+1,k+1]/length(var))*100
        end
    end
    return a
end
```

```julia
n=100
var1=rand(0:1,n)
var2=rand(0:1,n)
m=10
conect=rand((1:n),n,m)
yo=rand(0:5,n,m-1)
a=onoffp(var1,conect,yo,100,50)
b=onoffp(var2,conect,yo,100,50)
plot(a[1:100,1],a[1:100,3],a[1:100,1],b[1:100,3])
title("n=100, m=10")
ylabel("Porcentaje de variables activadas")
xlabel("Pasos")
```
