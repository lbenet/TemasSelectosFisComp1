##El rotor pateado
Del hamiltoneano de la forma: 
$$H=p^2/2 +V(x)\sum_{n=-\infty}^{\infty}{\delta (t-nT)}$$
se deduce que las ecuaciones de movimiento son:
$$\dot x=\frac{\partial H}{\partial p}=p$$
$$\dot p=-\frac{\partial H}{\partial x}=-V'(x)\cdot \sum_{n}{\delta(t-nT)}$$
Se puede aproximar por el siguiente sistema discreto.
$$p_{n+1}=p_n+F(x_n)$$
$$x_{n+1}=x_n+p_{n+1}$$


Intentemos primero como $F(x)=Asin(x)$. $A$ es un parámetro que influencia el grado de caos.

```julia
using PyPlot, Interact, Reactive, TaylorSeries
```

```julia
function evolC{T<:Float64}(vec::Array{T,1},A::T,n::Int64)
    
    x0=vec[1]
    p0=vec[2]
    
    x0=mod2pi(x0)
    x=[x0]
    p=[p0]
    
    for i=1:n
        p0+=A*sin(x0)
        x0=mod2pi(x0+p0)
        push!(x,x0)
        push!(p,p0)
    end
    return (x,p)
end
```

```julia
function evolT{T<:Float64}(vec::Array{T,1},A::T,n::Int64)
    
    x0=vec[1]
    p0=vec[2]
    
    x0=mod2pi(x0)
    p0=mod2pi(p0)
    
    x=[x0]
    p=[p0]
    
    for i=1:n
        p0=mod2pi(p0 + A*sin(x0))
        x0=mod2pi(x0+p0)
        push!(x,x0)
        push!(p,p0)
    end
    return (x,p)
end
```

### El exponente de Lyapunov

Se analiza como va cambiando la distancia entre dos trayectorias con condiciones iniciales muy cercanas, con una separación inicial $d_0=d(t_0)$, suponiendo que el comportamiento temporal de dicha distancia es exponencial:

$$d(t)=d_0e^{\lambda t}$$

Si llega a pasar que las trayectorias se separan por una distancia considerable respecto al sistema se debe reiniciar el proceso, de manera que la trayectoria secundaria vuelva a comenzar a una distancia $d_0$ de la trayectoria principal, sobre la recta de union.


```julia
function Distancia{T<:Float64}(vec0::Array{T,1},vec1::Array{T,1}, dimension::Int64) 
    #Se especifica la dimensión para que no repita el cálculo de esta en cada ciclo.
    
    suma=0.0
    for i=1:dimension
        dif=vec0[i]-vec1[i]
        suma += dif^2
    end
    return sqrt(suma)  
end
```

```julia
function LyapDis{T<:Float64}(Ecuacion::Function, vec0::Array{T,1}, vec1::Array{T,1}, A::T, L::T, N::Int64)
    #La longitud L es un parámetro basado en el sistema que define la másima separación aceptable.
    
    #vec = [x,p]
    vec0==vec1 && return println("¡Debes elegir condiciones iniciales distintas para cada trayectoria!")
    
    K=length(vec0)
    D0=Distancia(vec0,vec1,K)
    
    Lambdas = Float64[]
    
    for i=1:N
        
        evol0 = Ecuacion(vec0, A, 1)
        evol1 = Ecuacion(vec1, A, 1)
        
        vec0 = [evol0[1][2],evol0[2][2]]   #Esto es porque las funciones almacenan la 
        vec1 = [evol1[1][2],evol1[2][2]]   #condición inicial en la primera entrada.
        
        D = Distancia(vec0,vec1,K)
    
        λ=log(D/D0)
        push!(Lambdas, λ)
        
    end
    
    Suma=0
    Tamaño = length(Lambdas)
    for i=1:Tamaño
        Suma+=Lambdas[i]
    end
        
    Promedio=Suma/Tamaño
    Pasos=[i for i=1:Tamaño]
    
    return (Promedio, extrema(Lambdas), Lambdas, Pasos)
end
```

##Rotor pateado amortiguado (The Web Map)
Del hamiltoneano de la forma: 
$$H=\frac{1}{2}(p^2+\omega _0 ^2 x^2) - \frac{\omega _0 K}{T} \cos x \sum_{n=-\infty}^{\infty}{\delta (\frac{t}{T}-n)}$$
se deduce que la ecuación de movimiento es:
$$\ddot x+\omega _0 ^2 x= - \frac{\omega _0 K}{T} \sin x \sum_{n=-\infty}^{\infty}{\delta (\frac{t}{T}-n)}$$
Se puede aproximar por el siguiente sistema discreto.
$$u_{n+1}=(u_n +K\sin v_n) \cos \alpha + v_n \sin \alpha$$
$$v_{n+1}=-(u_n+K \sin v_n)\sin \alpha + v_n\cos \alpha$$
donde se introdujeron las variables adimensionales:
$$ u=\frac{\dot x}{\omega _0}, v=-x, \alpha=\omega _0 T$$
Usando numeros complejos se puede reescribir más facilmente
$$ z_{n+1}=(z_n+K\sin \frac{z-z^*}{2i})e^{i\alpha} $$

Los casos interesantes se dan cuando hay resonancia entre la frecuencia del oscilador y los pateos, lo cual se expresa como:

$$ \alpha=\omega _0 T=\frac{2\pi}{q} $$

Y los cásos más interesantes se dan cuando $q\ge3$

```julia
function Web(vec,K,q,n) #Condiciones inciales, factor caótico, factor de resonancia, número de iteraciones
    
    u0=vec[1]
    v0=vec[2]
    u=[u0]     
    v=[v0]
    for i=1:n
        u0=(u[i]+K*sin(v[i]))cos(2π/q)+v[i]*sin(2π/q)
        v0=-(u[i]+K*sin(v[i]))sin(2π/q)+v[i]*cos(2π/q)
        push!(u,u0)
        push!(v,v0)
    end
    (v,u)
end
```

```julia
function LyapWeb{T<:Float64}(vec0::Array{T,1}, vec1::Array{T,1}, K::T, q::Int64 ,L::T, N::Int64)
    #La longitud L es un parámetro basado en el sistema que define la másima separación aceptable.
    #vec = [x,p]
    
    vec0==vec1 && return println("¡Debes elegir condiciones iniciales distintas para cada trayectoria!")
    
    K=length(vec0)
    D0=Distancia(vec0,vec1,K)
    
    Lambdas=Float64[]

    if L==0.0
    
        for i=1:N

            evol0 = Web(vec0, K, q, 1)
            evol1 = Web(vec1, K, q, 1)

            vec0 = [evol0[1][2],evol0[2][2]]   #Esto es porque las funciones almacenan la 
            vec1 = [evol1[1][2],evol1[2][2]]   #condición inicial en la primera entrada.

            D = Distancia(vec0,vec1,K)

            λ=log(D/D0)/i
            push!(Lambdas, λ)
            
        end
        
    else
        
        pasos = 0
        
        for i=1:N

            evol0 = Web(vec0, K, q, 1)
            evol1 = Web(vec1, K, q, 1)

            vec0 = [evol0[1][2],evol0[2][2]]
            vec1 = [evol1[1][2],evol1[2][2]]

            D = Distancia(vec0,vec1,K)
            
            pasos += 1

            if D>L

                λ=log(D/D0)/pasos
                push!(Lambdas, λ)
                V=vec1-vec0
                vec1 = vec0 + D0*V/D
                pasos = 0
            end
        end
    end
    
    Suma=0
    Tamaño = length(Lambdas)
    if Tamaño == 0
        return println("Las trayectorias nunca se separaron más allá del parámetro establecido: ", L)
    else
        for i=1:Tamaño
            Suma+=Lambdas[i]
        end
    end

    Promedio=Suma/Tamaño
    Ext = extrema(Lambdas)
    Cortes = [n for n=1:Tamaño]

    return (Promedio, Ext , Lambdas, Cortes, Tamaño )
    
end
```

##Atractor variable

####Este sistema es de interés solamente porque presenta un comportamiento similar al de varios atractores distintos dependiendo de los valores que se asignen a sus parámetros;  sin ser equivalente topológicamente con ninguno de ellos. Puede parecerse al atractor de Lorentz ($a=2.4$ y $b=0.78$), al atractor de Chen ($a=1.5$ y $b=0.2$), al atractor de Lü($a=2.4$ y $b=0.28$) o a un nuevo atractor ($a=2.4$ y $b=0.12$).

####Las ecuaciónes que definen al sistema son:

$$\dot x=a(-x+y+yz)$$
$$\dot y=y-xz$$
$$\dot z=-bz+xy$$

```julia
const orderTaylor = 35;
const epsilon = 1.0e-35;
```

####Esta función determina el desarrollo de Taylor a partir de las condiciones iniciales. Esto lo hace recursivamente utilizando la siguiente regla, hasta el orden convenido para los polinomios de Taylor ($orderTaylor$).

$$ x_{[n+1]} = \frac{f_{[n]}(x,y,z,t)}{n+1} $$

####Donde $f_{[n]}(x,t)$ es el $n$-ésimo coeficiente de Taylor de $f(x,y,z,t)$, que define la ecuación de movimiento $\dot{x} = f(x,y,z,t)$. Así mismo $x_{[n]}$ es el $n$-ésimo coeficiente de Taylor para la variable dependiente $x(t)$, alrededor de  la condición inicial $x(t_0)=x_{[0]}$.

```julia
function Atractor{T<:Float64}(vector::Array{Taylor{T},1})
    
    (xT,yT,zT) = vector[1:3] #Nombrando las entradas, que son taylors
    
    ord = zT.order #Para utilizar el orden con el que entran los Taylor

    #Taylors de ceros donde se van almacenando las operaciones en cada entrada
    yzT = zeros( T, ord+1 )  #La multiplicación de la primera ecuación  xz
    xzT = zeros( T, ord+1 )  #La multiplicación de la segunda ecuación   xz
    xyT = zeros( T, ord+1 )  #La multiplicación de la tercera ecuación   xy

    for i = 1:ord
        
        # Implementación utilizando funciones predefinidas
        yzT[i] = TaylorSeries.mulHomogCoef(i-1, yT.coeffs, zT.coeffs)
        xzT[i] = TaylorSeries.mulHomogCoef(i-1, xT.coeffs, zT.coeffs)
        xyT[i] = TaylorSeries.mulHomogCoef(i-1, xT.coeffs, yT.coeffs)
        
        xT.coeffs[i+1] = a*(-xT.coeffs[i]+yT.coeffs[i]+yzT[i])/i
        yT.coeffs[i+1] = (yT.coeffs[i]-xzT[i])/i
        zT.coeffs[i+1] = (-b*zT.coeffs[i]+xyT[i])/i
    end
    
    return [Taylor(xT,ord),Taylor(yT,ord),Taylor(zT,ord)]
end
```

####Esta función determina el paso de tiempo óptimo para mantener el error debajo del permitido ($\epsilon, epsilon$) utilizando los dos últimos coeficientes del polinómio de taylor. Para esto se utiliza la siguiente regla:

$$h = \min[\, (\epsilon/x^{[p-1]})^{1/(p-1)}, (\epsilon/x^{[p]})^{1/p}\, ] $$

####donde $p$ el orden del polinomio de Taylor, $x^{[n]}$ es el coeficiente de Taylor de $n$-ésimo orden.

```julia
function saltoTemporal{T<:Float64}(Taylorazo::Taylor{T})
    
    ord=Taylorazo.order
    h = Inf
    for k in [ord-1, ord]
        kinv = 1.0/k
        aux = abs( Taylorazo.coeffs[k+1] )
        h = min(h, (epsilon/aux)^kinv)
    end
    return h
end
```

####Esta función utiliza a las dos anteriores para formar un taylor inicial a partir de las condiciones inciales,para después determinar el tamaño del paso necesario y calcular el valor del siguiente estado usando el método de Horner.

```julia
function unPasito{T<:Float64}(Equations::Function, vec0::Array{T,1})
    
    n = length(vec0)
    vec0T = [ Taylor([vec0[i]], orderTaylor) for i=1:n ]
    vec1T = Equations(vec0T)
    
    # Tamaño del paso temporal
    hh = Inf
    for i=1:n
        h1 = saltoTemporal( vec1T[i] )
        hh = min( hh, h1 )
    end
    
    # Evaluación tras el paso calculado
    for i=1:n
        vec0[i] = evalTaylor( vec1T[i], hh )  #evalTaylor() evalúa los polinomios de cada variable usando el método de Horner
    end
    
    # Obtenemos el nuevo vector y el tamaño del paso utilizado
    return (vec0,hh)
end
```

####Esta función itera a la anterior hasta que la suma de los pasos tomados alcanza un valor predeterminado, almacena todos los estados intermedios por los que pasa el sistema.

```julia
function evolAtrac{T<:Float64}(Equations::Function, vec0::Array{T,1}, tiempo::T)
    
    S=0.0         #Parámetro que frenará el ciclo while
    
    x=[vec0[1]]   #Arreglos donde se guardarán los valores de cada iteración
    y=[vec0[2]]
    z=[vec0[3]]
    t=[tiempo]
    
    while S<=tiempo
        
        eval = unPasito(Equations, vec0)
        vec0 = eval[1]
        paso = eval[2]
        
        push!(x,vec0[1])
        push!(y,vec0[2])
        push!(z,vec0[3])
        push!(t,paso)
             
        S += paso
        
    end
    return (x,y,z,t)
end
```

```julia
function GrafAtrac{T<:Float64}(vec0::Array{T,1}, tiempo::T)
    E = evolAtrac (Atractor, vec0, tiempo)
    return plot3D(E[1],E[2],E[3])
end
```

```julia
#Esta función es casi igual a stepSized pero incluye ambas trayectorias

function pasoLyap{T<:Float64}(Equations::Function, vec1::Array{T,1}, vec2::Array{T,1})
    
    n = length(vec1)
    vec1T = [ Taylor([vec1[i]], orderTaylor) for i=1:n ]
    vec2T = [ Taylor([vec2[i]], orderTaylor) for i=1:n ]
    VEC1T = Equations(vec1T)
    VEC2T = Equations(vec2T)
    
    hh = Inf
    for i=1:n
        h1 = saltoTemporal(VEC1T[i])
        h2 = saltoTemporal(VEC2T[i])
        hh = min( hh, h1 ,h2 )
    end
    
    for i=1:n    #evalTaylor() evalúa los polinomios de cada variable usando el método de Horner
        vec1[i] = evalTaylor( VEC1T[i], hh )  
        vec2[i] = evalTaylor( VEC2T[i], hh )  
    end
    
    return (vec1,vec2,hh)
end
```

```julia
function expLyap{T<:Float64}(Equations::Function, vec0::Array{T,1}, vec1::Array{T,1}, L::T, N::Int64)
    
    vec0==vec1 && return println("¡Debes elegir condiciones iniciales distintas para cada trayectoria!")
    
    K=length(vec0)
    D0=Distancia(vec0,vec1,K)
    Lambdas=Float64[]
    paso=0
    
    if L==0.0
    
        for i=1:N
            
            eval = pasoLyap(Equations, vec0, vec1)
            vec0 = eval[1]
            vec1 = eval[2]
            paso += eval[3]

            D = Distancia(vec0,vec1,K)
            λ=log(D/D0)/paso
            push!(Lambdas, λ)
            
        end
        
    else
        
        for i=1:N

            eval = pasoLyap(Equations, vec0, vec1)
            vec0 = eval[1]
            vec1 = eval[2]
            paso += eval[3]

            D = Distancia(vec0,vec1,K)

            if D>L
                λ=log(D/D0)/paso
                push!(Lambdas, λ)
                V=vec1-vec0
                vec1 = vec0 + D0*V/D
                paso = 0
            end
        end
        
    end
    
    Suma=0
    Tamaño = length(Lambdas)
    if Tamaño == 0
        return println("Las trayectorias nunca se separaron más allá del parámetro establecido: ", L)
    else
        for i=1:Tamaño
            Suma+=Lambdas[i]
        end
    end

    Promedio=Suma/Tamaño
    Ext = extrema(Lambdas)
    Cortes = [n for n=1:Tamaño]

    return (Promedio, Ext , Lambdas, Cortes, Tamaño )
    
end
    
```

#Péndulo invertido con una base que oscila verticalmente.

La forma de oscilación de la base es hamónica: $$y(t)=A\cos{\omega t}$$ $$\dot y (t)=-A\sin{\omega t}$$

La posición del péndulo de largo $l$ está dada por: $$(X,Y)=(l\sin{\theta}, y+l\cos{\theta})$$

Vemos entonces que el cuadrado de la velocidad es: $$V^2=\dot X ^2 +\dot Y^2=(l \dot \theta) ^2+\dot y -2l\dot y \dot \theta\sin{\theta}$$

Por lo que el lagrangiano queda dado por: $$\mathcal{L}=\frac{1}{2}m((l \dot \theta) ^2+\dot y -2l\dot y \dot \theta\sin{\theta}) - mg(y+l\cos{\theta})$$

La ecuación de Lagrange es: $$ \frac{d}{dt}(\frac{\partial \mathcal{L}}{\partial \dot \theta}) = \frac{\partial \mathcal{L}}{\partial \theta} \rightarrow l\ddot \theta-\ddot y \sin\theta=g\sin\theta \rightarrow l\ddot\theta +\sin\theta(A\omega^2\cos{\omega t}-g)=0$$

Si se define el momento geralizado: $$p_\theta \equiv p \equiv \frac{\partial \mathcal{L}}{\partial \dot \theta} = ml(l\dot \theta - \dot y \sin \theta)$$
de donde obtenemos: $$\dot \theta = \frac{p}{ml^2}+\frac{\dot y \sin \theta}{l} = \frac{1}{ml^2}(p+ml\dot y \ sin \theta)$$

Ahora obtenemos el hamiltoneano: $$\mathcal{H} = p\dot\theta - \mathcal{L} = \frac{1}{2ml^2} (p+ml \dot y \sin \theta)^2 - \frac{1}{2} m \dot y + mg(y+l\cos\theta)$$

Las ecuaciónes de Hamilton que se obtienen a partir de este son:
$$\dot \theta = \frac{1}{ml^2}(p+ml\dot y \ sin \theta)$$
$$\dot p = -\frac{p\dot y \cos \theta}{l} -m \dot y^2 \sin \theta \cos \theta +  mgl\sin \theta $$
$$\dot t = 1 $$


```julia
#Esta función toma el vector de condiciones iniciales pero en su versión con desarrollos de taylor 
#en cada entrada a partir de estas.


function penInv{T<:Float64}(vector::Array{Taylor{T},1})   #Las constantes son m,l,A,ω,g
 
    (θT,pT,tT)=vector[1:3]
    
    ord=θT.order  #Podemos usar el orden de θT en particular porque todos son iguales.
    for i=1:ord 
        
        θTt = Taylor(θT.coeffs[1:i],i-1)
        pTt = Taylor(pT.coeffs[1:i],i-1)
        tTt = Taylor(tT.coeffs[1:i],i-1)
        
        sθ = sin(θTt)  #Elementos auxiliares
        cθ = cos(θTt)
        #y = A*cos(ω*tTt)
        #yDot = -A*ω*sin(ω*tTt)
        y = A*sin(ω*tTt)
        yDot = A*ω*cos(ω*tTt)

        
        θDot = (pTt+m*l*yDot*sθ)/(m*l^2)
        pDot = -pTt*yDot*cθ/l - m*yDot^2*sθ*cθ + m*g*l*sθ
        tDot = Taylor(1.0,i-1)
        
        θT.coeffs[i+1] = θDot.coeffs[i]/i
        pT.coeffs[i+1] = pDot.coeffs[i]/i
        tT.coeffs[i+1] = tDot.coeffs[i]/i
        
    end
    return[Taylor(θT,ord),Taylor(pT,ord),Taylor(tT,ord)]
end
```

```julia
function evolInv{T<:Float64}(vec0::Array{T,1}, tiempo::T)
    
    t0=vec0[3]         #Tiempo inicial
    
    θ=[vec0[1]]        #Arreglos donde se guardarán los valores de cada iteración
    p=[vec0[2]]
    t=[vec0[3]]
    y=[A] 
    
    while t0<=tiempo
        
        eval = unPasito(penInv, vec0)
        vec0 = eval[1]
        paso = eval[2]
        
        t0=vec0[3]
                
        push!(θ,vec0[1])
        push!(p,vec0[2])
        push!(t,vec0[3])
        push!(y,A*cos(ω*t0))
        
    end
    
    return (θ,p,t,y)
end
```

```julia
function enerInv{T<:Float64}(θV::Array{T,1},pV::Array{T,1},tiempos::Array{T,1})
    
    Energias=Float64[]
    n=length(θV)
    
    for i=1:n
        θ = θV[i]
        p = pV[i]
        t = tiempos[i]
        

        y = A*cos(ω*t)
        yDot = -A*sin(ω*t)
        H = (1/(2*m*(l^2)))*(p+m*l*yDot*sin(θ))^2 - (1/2)*(m*yDot) + m*g*(y+l*cos(θ))
        push!(Energias,H)
    end
    
    relH = Float64[]
    #div = Float64[]
    for k=1:n
        relativo = (Energias[k]-Energias[1])/Energias[1]
        push!(relH, relativo)
    end
        

    return plot(tiempos, relH,"b")
end
```

#El péndulo doble

El hamiltoneano es: $$\mathcal{H} =\frac{l_2^2 m_2 p_{\theta, 1}^2-2 l_1 l_2 m_2 \cos(\theta_1-\theta_2) p_{\theta, 1} p_{\theta, 2}+l_1^2 (m_1+m_2) p_{\theta, 2}^2}{2 l_1^2 l_2^2 m_2 (m_1 + m_2 \sin^2({\theta_1-\theta_2})))} - g l_1 (m_1+m_2) \cos\theta_1-g l_2 m_2 \cos\theta_2$$
Las ecuaciones de movimiento son:
$$\dot \theta_1 = \frac{\partial \mathcal{H}}{\partial p_{\theta_1}} = \frac{l_2 p_{\theta_1} - l_1 p_{\theta_2} \cos(\theta_1 -\theta_2)}{l_1^2 l_2[m_1 + m_2 \sin^2(\theta_1-\theta_2)]}$$

$$\dot \theta_2 = \frac{\partial \mathcal{H}}{\partial p_{\theta_2}} = \frac{l_1 (m_1+m_2)p_{\theta_2}-m_2 l_2 p_{\theta_1}\cos(\theta_1-\theta_2)}{l_1 l_2^2 m_2[m_1 + m_2\sin^2(\theta_1-\theta_2)]}$$

$$\dot p_{\theta_1} = -\frac{\partial \mathcal{H}}{\theta_1} = -(m_1+m_2)g l_1 \sin \theta_1 - C_1 + C_2$$

$$\dot p_{\theta_2} = -\frac{\partial \mathcal{H}}{\theta_2} =  -m_2 g l_2 \sin \theta_2 +C_1 - C_2$$

$$C_1 \equiv \frac{p_{\theta_1}p_{\theta_2}\sin(\theta_1-\theta_2)}{l_1 l_2[m_1 + m_2 \sin^2(\theta_1-\theta_2)]} $$

$$C_2 \equiv \frac{ l_2^2m_2p^2_{\theta_1} + l_1^2 p^2_{\theta_2}(m_1+m_2) - l_1 l_2 m_2 p_{\theta_1}p_{\theta_2} \cos(\theta_1-\theta_2)} { 2l_1^2 l_2^2[m_1 + m_2 \sin^2(\theta_1-\theta_2)]^2 } \sin[2(\theta_1-\theta_2)] $$

```julia
#Esta función toma el vector de condiciones iniciales pero en su versión con desarrollos de taylor 
#en cada entrada a partir de estas.


function penDob{T<:Float64}(vector::Array{Taylor{T},1})   #Las constantes son m1,m2,l1,l2,g
 
    (θ1T,θ2T,p1T,p2T)=vector[1:4]
    #@show(m1,m2)
    ord=θ1T.order  #Podemos usar el orden de θT en particular porque todos son iguales.
    for i=1:ord 
        
        θ1Tt = Taylor(θ1T.coeffs[1:i],i-1)
        θ2Tt = Taylor(θ2T.coeffs[1:i],i-1)
        p1Tt = Taylor(p1T.coeffs[1:i],i-1)
        p2Tt = Taylor(p2T.coeffs[1:i],i-1)
        
        θ12 = θ1Tt-θ2Tt
        s12 = sin(θ12)  #Elementos auxiliares
        c12 = cos(θ12)
        den = m1+m2*s12^2
        p12 = p1Tt^2
        p22 = p2Tt^2
        p1p2 = p1Tt*p2Tt
        
        X = m2*p12*l2^2
        Y = p22*(m1+m2)*l1^2
        Z = l1*l2*m2*p1p2*c12
        
        C1 = p1p2*s12/(l1*l2*den)
        C2 = (X+Y-Z)*sin(2*θ12)/(2*(l1*l2*den)^2)
        
        θ1Dot = (l2*p1Tt-l1*p2Tt*c12)/(l2*den*l1^2)
        θ2Dot = (l1*(m1+m2)*p2Tt - l2*m2*p1Tt*c12)/(l1*l2^2*m2*den)
        p1Dot = -(m1+m2)*g*l1*sin(θ1Tt) - C1 + C2
        p2Dot = -m2*g*l2*sin(θ2Tt) + C1 - C2
        
        θ1T.coeffs[i+1] = θ1Dot.coeffs[i]/(i)
        θ2T.coeffs[i+1] = θ2Dot.coeffs[i]/(i)
        p1T.coeffs[i+1] = p1Dot.coeffs[i]/(i)
        p2T.coeffs[i+1] = p2Dot.coeffs[i]/(i)
        
    end
    return[Taylor(θ1T,ord),Taylor(θ2T,ord),Taylor(p1T,ord),Taylor(p2T,ord)]
end
```

```julia
function evolDob{T<:Float64}(vec0::Array{T,1}, tiempo::T)
    
    S=0.0         #Parámetro que frenará el ciclo while
    
    θ1=[vec0[1]]   #Arreglos donde se guardarán los valores de cada iteración
    θ2=[vec0[2]]
    p1=[vec0[3]]
    p2=[vec0[4]]
    
    tiempos = [S]   #Esto lo usaremos para analizar la energía
    
    while S<=tiempo
        
        eval = unPasito(penDob, vec0)
        vec0 = eval[1]
        step = eval[2]
        
        push!(θ1,vec0[1])
        push!(θ2,vec0[2])
        push!(p1,vec0[3])
        push!(p2,vec0[4])
                
        S += step
        
        push!(tiempos,S)
        
    end
    
    return (θ1,θ2,p1,p2,tiempos)
end
```

```julia
function enerDob{T<:Float64}(θ1V::Array{T,1},θ2V::Array{T,1},p1V::Array{T,1},p2V::Array{T,1},tiempos::Array{T,1})
    
    #(Pasos:Array{Array{T,1},1})
    #(θ1V::Array{T,1},θ2V::Array{T,1},p1V::Array{T,1},p2V::Array{T,1},tiempos::Array{T,1})
    
    Energias=Float64[]
    n=length(θ1V)
    
    #=
    θ1V=Pasos[1]
    θ2V=Pasos[2]
    p1V=Pasos[3]
    p2V=Pasos[4]
    tiempos = Pasos[5]
    =#
    
    for i=1:n
        
        θ1 = θ1V[i]
        θ2 = θ2V[i]
        p1 = p1V[i]
        p2 = p2V[i]
        
        
        #Se calcula el hamiltoneano porque representa la energía total.
        θ12=θ1-θ2
        A=(l2^2)*m2*(p1^2)
        B=(l1^2)*(m1+m2)*(p2^2)
        C=2*l1*l2*m2*p1*p2*cos(θ12)
        D=2*(l1^2)*(l2^2)*m2
        E=m1+m2*(sin(θ12))^2
        F=g*l1*(m1+m2)*cos(θ1)
        G=g*l2*m2*cos(θ2)
        
        H=((A+B-C)/(D*E)) - F - G
        
        
        #H = ((l2^2*m2*p1^2-2*l1*l2*m2*cos(θ1-θ2)*p1*p2+l1^2*(m1+m2)*p2^2)/(2*l1^2*l2^2*m2*(m1+m2*(sin(θ1-θ2))^2)))-g*l1*(m1+m2)*cos(θ1) -g*l2*m2*cos(θ2)
        push!(Energias,H)
    end
    
    #=
    relH = Float64[]
    div = Float64[]
    for k=1:n
        relativo = (Energias[k]-Energias[1])/Energias[1]
        if relativo > 10^10000
            push!(div,k)
        end
        push!(relH, relativo)
    end
    =#
        
    #return plot(tiempos,Energias,"blue",tiempos,relH,"green")
    #return plot(tiempos,relH,"green")
    return plot(tiempos,Energias,"b.")
    #return div
end
```
