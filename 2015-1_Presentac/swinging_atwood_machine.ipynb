#Swinging Atwood's Machine

Éste modelo es muy parecido al la máquina de Atwood convencional. La diferencia es que uno de los lados de la polea tiene la libertad de rotar sin colisionar con la otra cuerda.

Es un sistema con dos grados de libertad. Se pueden derivar sus ecuaciones de movimiento con el sistema **Hamiltoniano** o con el sistema **Lagrangiano** .
Éste es un sistema de masas $m$, la masa que se columpia y $M$, la masa que no se columpia. Se tiene que la energía cinética del sistema es: 

   $$  T = \frac {1}{2} M v_M^2 + \frac {1}{2} m v_m^2 = \frac {1}{2} M \dot {r^2} + \frac {1}{2} m (\dot {r^2} + r^2 \dot{\theta ^ 2})  $$ 
   
con $r$ la distancia de $m$ a su pivote de giro y $\theta$ el ángulo de giro de $m$.

Se tiene que la **Energía Potencial** $\ U$ del sistema es:

  $$ U = Mgr - mgr\cos{\theta} $$
Así, podemos obtener el **Hamiltoniano** y el **Lagrangiano** del sistema: 

$$\ {\mathcal{L}} = T - U = \frac {1}{2} M \dot {r^2} + \frac {1}{2} m (\dot {r^2} + r^2 \dot{\theta ^2}) - Mgr + mgr\cos{\theta} $$
<p>
$$\ {\mathcal{H}} = T + U = \frac {1}{2} M \dot {r^2} + \frac {1}{2} m (\dot {r^2} + r^2 \dot{\theta ^2}) + Mgr - mgr\cos{\theta} $$

Esto se puede expresar en términos de los **momentos** canónicos $p_r$ y $p_{\theta}$, donde:

   $$p_r =\frac {\partial {\mathcal{L}}}{\partial \dot{r}}=\frac {\partial {T}}{\partial \dot{r}} = (M + m)\dot{r} ...(1)$$
   <p>
   $$p_\theta =\frac {\partial {\mathcal{L}}}{\partial \dot{\theta}}=\frac {\partial {T}}{\partial \dot{\theta}} = mr^2\dot \theta ...(2)$$ 
   <p><p>
   $$ \therefore \mathcal{H} = \frac {p_r^2}{2(M+m)} + \frac {p_\theta^2}{2mr^2} + Mgr - mgr\cos{\theta} $$
    
Haciendo un análisis del **Lagrangiano** para $r$ y $\theta$ se pueden obtener una serie de *ecuaciones diferenciales acopladas* que serán con las cuales trabajaremos en este proyecto. 
Notemos que, para $\theta$:

   $$ \frac {\partial {\mathcal{L}}}{\partial {\theta} } = \frac{d}{dt} (\frac {\partial {\mathcal{L}}}{\partial \dot{\theta}}) \implies -mgr\sin{\theta} = 2mr\dot{r}\dot{\theta} + mr^2\ddot{\theta} $$
   
Para $r$ : 

   $$ \frac {\partial {\mathcal{L}}}{\partial {r} } = \frac{d}{dt} (\frac {\partial {\mathcal{L}}}{\partial \dot{r}}) \implies mr\dot{\theta^2} - Mg + mg\cos{\theta} =(m+M)\ddot{r}$$ 
   
Para simplificar las expresiones, definimos un **cociente de masa** $\mu$ como $\mu := \frac{M}{m}$.

Finalmente nos quedan expresadas las siguientes *ecuaciones diferenciales acopladas*, que determinan el movimiento del sistema, donde:


###Ecuaciones de Movimiento: 

#### $$\dot{r} = \frac {p_r}{M + m}$$

#### $$\dot{p_r} = \frac {p_{\theta}^2}{mr^3} - Mg + mg\cos{\theta} $$ 

#### $$\dot{\theta} = \frac {p_\theta}{mr^2} $$

#### $$\dot{p_\theta} = -mgr\sin{\theta} $$


Una vez que se tienen las ecuaciones de movimiento, se emplea algún método numérico para resolverlas.
En este proyecto se utilizó el método de Taylor, donde, dadas las *condiciones iniciales* del sistema, se hace un desarrollo en serie de potencias para cada una de las ecuaciones diferenciales y se hace un proceso iterativo donde, para cada iteración, se calcula un *paso* o *StepSize* temporal para el cual se vuelven a calcular los coeficientes en ese momento.

Se incluye enseguida el código que implementa la variable **tipo** Taylor, donde se define el polinomio de Taylor y las operaciones con éste.

Imponemos las constantes del sistema.
$\epsilon$ es el error mínimo permitido en nuestro desarrollo en series, $ordermax$ es el orden del Polinomio al que llegaremos.
$g$ es la constante de gravitación aproximada.

```julia
const ordermax = 25
const epsilon = 1.0e-25
const g = 9.81
```

A continuación se define una función *GetCoeffs*, donde metemos las ecuaciones de movimiento y usamos la fórmula iterativa 
$$x_{n+1} = \frac {f_n}{n+1} \ \ \ ...(1)$$

```julia
function GetCoeffs{T<:Number}(cond_init::Array{Taylor{T}},M::T=5.0,m::T=1.0) #Damos un Taylor de orden 0 con orden=ordermax para cada variable
    (rT,p_rT,thT,p_thT) = cond_init[1:4] #Identificamos cada taylor con cada entrada de la tupla
    
    ord = rT.order #Puede ser cualquiera (rT,p_rT,thT,p_thT) ya que todas tienen el mismo orden
    for i in 0:ord-1
        i_next = i+1
        
        # Hacemos la expansion hasta orden i
        rTt    = Taylor(rT.coeffs[1:i+1],i)
        p_rTt  = Taylor(p_rT.coeffs[1:i+1],i)
        thTt   = Taylor(thT.coeffs[1:i+1],i)
        p_thTt = Taylor(p_thT.coeffs[1:i+1],i)
        
        # Definimos la f(x), en este caso, las ecuaciones del sistema
        # NOTA: Aquí estamos aprovechando que previamente habíamos definido las operaciones con Taylor
        r_dot   = 1/(M+m)*p_rTt
        aux     = p_thTt/(rTt*rTt) #la definimos para que el ciclo tenga que hacer menos cálculos y tarde menos.
        pr_dot  = 1/m * aux * aux * rTt -M*g + m*g*cos(thTt)
        th_dot  = 1/m * aux
        pth_dot = -m*g*rTt*sin(thTt)
        
        #Ahora sí, usamos la expresión (1) para calcular los coeficientes
        rT.coeffs[i_next+1]    = r_dot.coeffs[i_next]   / i_next    #le va metiendo los coeficientes a los zeros de rT
        p_rT.coeffs[i_next+1]  = pr_dot.coeffs[i_next]  / i_next
        thT.coeffs[i_next+1]   = th_dot.coeffs[i_next]  / i_next
        p_thT.coeffs[i_next+1] = pth_dot.coeffs[i_next] / i_next 
    end
    
    return Coord = [ Taylor(rT, ord),Taylor(p_rT, ord),Taylor(thT, ord),Taylor(p_thT, ord)]
end
```

Una vez encontrados los coeficientes, podemos calcular el paso para cada iteración de estos, donde vimos que para tener una $h$ que sea realista, necesitamos que: 
$$\epsilon = x_n h^n \implies h = {(\frac{\epsilon}{x_n})} ^{\frac{1}{n}} $$

Para tener mejores posibilidades de tener buenas cuentas, tomaremos: 
$$ h = \min[\, (\epsilon/x_{n-1]})^{1/(n-1)}, (\epsilon/x_{n})^{1/n}\, ] $$


```julia
#Nos regresa el brinco más grande posible entre los últimos dos órdenes del Taylor a un momento dado.
function StepSize{T<:Real}(x::Taylor{T},eps::T=epsilon)
    ord = x.order
    h= Inf
    for k in ord-1:ord
        kinv=1/k
        aux = abs(x.coeffs[k+1])
        h = min(h, (eps/aux)^kinv)
    end
    return h
end
```

```julia
@time StepSize(l,1.0e-16) #epsilon ya está como default
```

Ahora que tenemos el paso para el primer Taylor, debemos hacer una función que empleé este *StepSize* y calcule, para un tiempo "stepsize" después, los coeficientes de Taylor en el tiempo nuevo.

Para esto, se utiliza el método de horner para evaluar a los polinomios obtenidos, ya que en clase se vio que este método es más eficiente que evaluar todo el polinomio *a lo wey*.

```julia
#para hacer el polinomio evaluado para alguna t y un taylor a
function evalHorner{T<:Number}(a::Taylor{T},h::T)
    A = a.coeffs
    ord = a.order+1
    coeff_t=A[end]
    for i in ord-1:-1:1
        coeff_t = A[i] + h*coeff_t     
    end
    return coeff_t
end
```

```julia
@time evalHorner(Taylor([3.0,2.0],ordermax),0.5)
```

*CoordVecs* es la implementación fuerte del código. Aquí se utilizan todas las funciones definidas anteriormente para, dadas condiciones iniciales, un error mínimo y un orden máximo, se calcule la evolución temporal de nuestro sistema dinámico; en este caso, la **Swinging Atwood Machine**

```julia
function CoordVecs{T<:Number}(CI::Array{Taylor{T}},t_tot::T,eps::T=epsilon,M=10.0,m=1.0)
   
    t = 0.0
    t_ar = [0.0]
    ord = CI[1].order
    # Tomamos las condiciones iniciales como floats
    Coeffs_R   = CI[1].coeffs[1]
    Coeffs_Pr  = CI[2].coeffs[1]
    Coeffs_Th  = CI[3].coeffs[1]
    Coeffs_Pth = CI[4].coeffs[1]
    
    # Hagamos un array de las condiciones iniciales, a estas les vamos a ir agragando los valores a todos
    #los tiempos posteriores, de modo que podamos graficar los resultados.
    X  = [CI[1].coeffs[1]]
    Px = [CI[2].coeffs[1]]
    Y  = [CI[3].coeffs[1]]
    Py = [CI[4].coeffs[1]]
    
    # El primer evalHorner debe un Taylor de las CI, así que todas las h_'s deben ser 0.
    h_R   = 0.0
    h_Pr  = 0.0
    h_Th  = 0.0
    h_Pth = 0.0
    
    while t <= t_tot
        
        # Encontramos todos los coeficientes de Taylor
        GC = GetCoeffs([Taylor(Coeffs_R,ord),Taylor(Coeffs_Pr,ord),Taylor(Coeffs_Th,ord),Taylor(Coeffs_Pth,ord)],M,m)
        Coeffs_R   = GC[1]
        Coeffs_Pr  = GC[2]
        Coeffs_Th  = GC[3]
        Coeffs_Pth = GC[4]
        
        # Con los coeficientes, definimos el step para ese momento
        h_R    = StepSize(Coeffs_R,eps)
        h_Pr   = StepSize(Coeffs_Pr,eps)
        h_Th   = StepSize(Coeffs_Th,eps)
        h_Pth  = StepSize(Coeffs_Pth,eps)
         
        h = min(h_R,h_Pr,h_Th,h_Pth)
        
        # Evaluamos el polinomio al tiempo h_i y se imponen unas nuevas condiciones iniciales
        Coeffs_R   = evalHorner(Taylor(Coeffs_R),h)
        Coeffs_Pr  = evalHorner(Taylor(Coeffs_Pr),h)
        Coeffs_Th  = evalHorner(Taylor(Coeffs_Th),h)
        Coeffs_Pth = evalHorner(Taylor(Coeffs_Pth),h)
        
        #ver cuando el punto cruza la condición que yo impongo bajo una cierta energía
        
        # Armamos, para c/u de las Eqs, un arreglo con sus puntos a cada step de tiempo posterior.
        # Imponemos cuánto tiempo ha pasado en total
        X  = push!(X,Coeffs_R)
        Px = push!(Px,Coeffs_Pr)
        Y  = push!(Y,Coeffs_Th)
        Py = push!(Py,Coeffs_Pth)
        
        t    = h + t
        t_ar = push!(t_ar,t)
        #print("\n Coeffs_Pth = $Coeffs_Pth, \ t = $t ")
    end 
    return (X,Px,Y,Py) , t_ar 
end
```

Para ver la eficiencia del método de Taylor lo mejor es compararlo con algún otro método que se considere fiable.
Emplearemos entonces el método de **Runge-Kutta** para el problema de la **Swinging Atwood's Machine**, donde: 

####Runge-Kutta 4:
$$y_{n+1}=y_n+\frac{h}{6}(k_1+2k_2+2k_3+k4)$$
$$t_{n+1}=t_n+h$$
Con:
$$k_1=f(t_n,y_n)$$
$$k_2=f(t_n+\frac{h}{2},y_n+\frac{h}{2}k_1)$$
$$k_3=f(t_n+\frac{h}{2},y_n+\frac{h}{2}k_2)$$
$$k_4=f(t_n+h,y_n+h*k_3)$$

Recordemos que en la **Máquina de Atwood**, las ecuaciones diferenciales correspondientes son:

####Ecuaciones de Movimiento: 
$$\dot{r}   = \frac {p_r}{M + m}$$
$$\dot{p_r} = \frac {p_{\theta}^2}{mr^3} - Mg + mg\cos{\theta} $$ 
$$\dot{\theta} = \frac {p_\theta}{mr^2} $$
$$\dot{p_\theta} = -mgr\sin{\theta} $$


```julia
function Rungekutta4(CI,f,t_inic,h,K=[3.0,1.0,9.81])
    #K es un arreglo de constantes del problema (en el swinging sería la M,m,g) 
    K1 = f(t_inic,CI,K)
    K2 = f(t_inic + h/2,CI + h/2*K1,K)
    K3 = f(t_inic + h/2,CI + h/2*K2,K)
    K4 = f(t_inic + h/2,CI + h/2*K3,K)
    C_paso = CI + h/6*(K1 + 2*K2 + 2*K3 + K4)
    C_paso
end
```

```julia
#Definimos una vez más las ecuaciones de movimiento
function Swinging{T<:Float64}(t::T,CI::Array{T,1},K::Array{T,1})
    R0   = CI[1]
    Pr0  = CI[2]
    Th0  = CI[3]
    Pth0 = CI[4]
    μ = 1/(K[1]+K[2])
    M = K[1]
    m = K[2]
    g = K[3]
    
    R_dot   = Pr0*μ
    Pr_dot  = Pth0^2/(m*R0^3) - M*g + m*g*cos(Th0)
    Th_dot  = Pth0/(m*R0^2)
    Pth_dot = -m*g*R0*sin(Th0)
    
    [R_dot, Pr_dot, Th_dot, Pth_dot]
end
```

Una vez definido el método y las ecuaciones de movimiento, se debe definir una función que agarre a dichas ecuaciones de movimiento y las ponga a iterar con el método deseado   

```julia
function integrar(CI,tf,h,f,K=[3.0,1.0,9.81])
    Ci = Vector{Float64}[]
    t = Float64[]
    for ti in 0:h:tf
        push!(Ci,CI)
        CI=Rungekutta4(CI,f,ti,h,K)
        push!(t,ti)
    end
    hcat(Ci...)',t
end
```

Ya listos ambos métodos, imponemos las condiciones iniciales del sistema, y dejamos que el código haga su trabajo.

```julia
th_0  = Float64[pi/2]
r_0   = 1.0
pr_0  = 0.0
pth_0 = 0.0

T_r   = Taylor(r_0,ordermax)
T_pr  = Taylor(pr_0,ordermax)
T_th  = Taylor(th_0,ordermax)
T_pth = Taylor(pth_0,ordermax)
CI    = [r_0,pr_0,th_0,pth_0]

M     = 16.0
m     = 1.0
t_tot = 75.0
CI
```

```julia
@time Ci,t = integrar(CI,t_tot,0.00005,Swinging,[M,m,g])
@time (R,Pr,Th,Pth),T = CoordVecs([T_r,T_pr,T_th,T_pth],t_tot,epsilon,M,m)

#le damos nombres a las variables de Runge-Kutta
pr  = Ci[1:end,2];
th  = Ci[1:end,3];
pth = Ci[1:end,4];
r   = Ci[1:end,1];
#Nos gustaría saber también el periodo de oscilación de nuestra máquina de Atwood
@time τ = 2*pi*sqrt(R/g);
@time τ_k = 2*pi*sqrt(r/g);
```

En este problema, tenemos un sistema cerrado; es decir, la energía total del sistema es constante.
<p>
Para comprobar la validez de estos métodos, es buena idea hacer un análisis energético para ver si se conserva la energía con estos métodos.
<p>
Tenemos que:

$$\mathcal{H} = E = T + U $$
en donde, por la simetría del problema, nos conviene trabajar en polares, de modo que:
<p>
$$ T = \frac {1}{2} M v_M^2 + \frac {1}{2} m v_m^2 = \frac {1}{2} M \dot {r^2} + \frac {1}{2} m (\dot {r^2} + r \dot{\theta})$$    
$$ U = Mgr - mgr\cos{\theta} $$

Donde en consideración de las ecuaciones (1) y (2), podemos reeplantear la **energía total** del sistema como
<p><p>
$$ E = \frac {p_r^2}{2(M+m)} + \frac {p_\theta^2}{2mr^2} + Mgr - mgr\cos{\theta} $$
constante en el tiempo.


```julia
#Se define la energía cinética
function kinetic(M,m,Pr,Pth,R,Th)
    n=length(Pr)
    μ=1/(2(M+m))
    K=[Pr[i]^2*μ+(Pth[i]^2)/(2m*(R[i]^2)) for i in 1:n]
end
```

```julia
#Se define la energía potencial
function pot(M,m,Pr,Pth,R,Th)
    n=length(Pr)
    μ=1/(2(M+m))
    U=[M*g*R[i]-m*g*R[i]*cos(Th[i]) for i in 1:n]
end
```

Con las anteriores, se calcula la energía total y la diferencia de energía *dE*, el cual presenta el cambio en la energía en términos de la épsilon que maneja la computadora

```julia
function Energy(M,m,Pr,Pth,R,Th)
    K=kinetic(M,m,Pr,Pth,R,Th)
    U=pot(M,m,Pr,Pth,R,Th)
    E=K+U
    dE=(E-E[1])/eps(E[1])
    E,U,K,dE
end
```

```julia
E,U,K,dE=Energy(M,m,Pr,Pth,R,Th);
e,u,k,de=Energy(M,m,pr,pth,r,th);

plot(τ_k,e)
plot(τ,E)
```

```julia
#Observemos el cambio de energía del método de Taylor en unidades de epsilon.
plot(T,dE,"green")
eps(E[1])
```

La energía tiene pequeñísimas variaciones del orden de $10^{-15}$ unidades, las cuales oscilan alrededor del cero.
Esto indica que no hay una *fricción computacional* que altere la energía, por lo que este método resulta muy efectivo.
