# Condensados de Bose Einstein:

## 2 pozos:

$$ \hat{H} = \epsilon_{1} \hat{a}_{1}^{\dagger} \hat{a}_{1} +  \epsilon_{2} \hat{a}_{2}^{\dagger} \hat{a}_{2} + \frac{E_{c}}{2} (\hat{a}_{1}^{\dagger 2} \hat{a}_{1}^{2} 
+ \hat{a}_{2}^{\dagger 2} \hat{a}_{2}^{2}) - E_{j} ( \hat{a}_{1}^{\dagger} \hat{a}_{2} + \hat{a}_{2}^{\dagger} \hat{a}_{1}) $$

Con:   $$\hat{n}_{i} = \hat{a}_{i}^{\dagger} \hat{a}_{i} $$
En el limite semi-clásico: $$\hat{a}_{i} \rightarrow I_{i}^{\frac{1}{2}} e^{i\phi_{j}} $$  
$$ \hat{a}_{i}^{\dagger} \rightarrow  I_{i}^{\frac{1}{2}} e^{-i\phi_{j}} $$

$$\Longrightarrow \hat{n}_{i} = I_{i} $$


$$ \hat{H} = \epsilon_{1} \hat{n}_{1} +  \epsilon_{2} \hat{n}_{2} + \frac{E_{c}}{2} (I_{1}(I_{1}-1) +  I_{2}(I_{2}-1))
- E_{j} ( I_{1}^{\frac{1}{2}}I_{2}^{\frac{1}{2}}Cos(\phi_{1}-\phi_{2})) $$

Usando: $$ W = J(\phi_{1} - \phi_{2}) + K\phi_{2} $$

$$I_{1} = \frac{\delta W}{\delta\phi_{1}} = J $$
$$I_{2} = \frac{\delta W}{\delta\phi_{2}} = K-J$$

Reemplacemos: $$\psi = \phi_{1} - \phi_{2}$$

$$  \therefore \hat{H} = \epsilon_{1} J +  \epsilon_{2} (K-J) + \frac{E_{c}}{2} (4J-2K)
- E_{j}(J(K-J))^{\frac{1}{2}} Cos(\psi) $$

Derivamos:
$$\dot J = -\frac{\delta H}{\delta\psi} = -2E_{j}\sqrt{J(K-J)} Sin(\psi)$$ 
$$\dot \phi = \frac{\delta H}{\delta J} = \epsilon_{1} - \epsilon_{2} + E_{c}[2J-K]-E_{j}[\frac{(K - 2J)}{2\sqrt{J(K-J)}}]Cos(\psi)$$ 


Siendo: J el número de partículas en el pozo izquierdo, K el número total de partículas, $\psi$ la diferencia de fase de la función de onda, $E_{j}$ la energia asociada al tunneling, $E_{c}$ la energia asociada a los choques entre partículas, $\epsilon_{1}$ y $\epsilon_{2}$ valores asociados con la altura de los pozos. 

## Integrador:

```julia
using TaylorSeries
```

```julia
const ordenTaylor = 18
const EpsAbs = 1.0e-20
```

```julia
#Constantes del modelo
e1 = 0.0
e2 = 0.0
Ec = 0.32/1170.0
Ej = 1.0

K = 100
```

```julia
function eqsBEC(xT::Array{Taylor{Float64},1})
    T = eltype(xT)
    psiT = xT[1]
    JT = xT[2]
    
    aux1 = sqrt( JT*(K-JT) )
    aux2 = K - 2JT
    tj = aux2/aux1
    aux3 = tj * cos(psiT)
    psiDotT = e1 - e2 - Ec * aux2 - Ej * aux3 / 2
    JDotT = - Ej * aux1 * sin(psiT)
    
    return psiDotT, JDotT
end
```

```julia
function Jet2BEC(vec0::Array{Float64,1}, eqsMotion::Function)
    psiCoef = [vec0[1]]
    JCoef   = [vec0[2]]
    
    # The jet of coefficients
    for k=0:ordenTaylor-1
        knext = k+1
        
        # Taylor expansions of order k
        psiT = Taylor(psiCoef, k)
        JT   = Taylor(JCoef, k)
        vecT = [psiT, JT]
        # Eqs of motion
        psiDotT, JDotT = eqsMotion(vecT)
        
        # Recursion
        aux = psiDotT.coeffs[knext] / knext
        push!(psiCoef, aux)
        aux = JDotT.coeffs[knext] / knext
        push!(JCoef, aux)
    end

    # Returned vector with all the Taylor polynomials
    return Taylor(psiCoef), Taylor(JCoef)
end
```

```julia
function stepsize(x::Taylor)
    ord = x.order
    h = Inf
    for k in (ord-1, ord)
        kinv = inv(k)
        aux = abs( x.coeffs[k+1] )
        h = min(h, (EpsAbs/aux)^kinv)
    end
    return h
end
```

```julia
function taylorInt( time, vec0, eqsMotion::Function )
    vec1T = Jet2BEC( vec0, eqsMotion )
    
    # The step-size
    hh = Inf
    for i=1:2
        h1 = stepsize( vec1T[i] )
        hh = min( hh, h1 )
    end
    hh < 0.0 && error("hh")
    
    # Values at t0+hh
    vec1 = zero(vec0)
    aux = evalTaylor(vec1T[1], hh)
    vec1[1] = aux #mod2pi(aux) 
    vec1[2] = evalTaylor(vec1T[2], hh)
    
    return hh, vec1
end
```

```julia
function HBEC(vec0::Array{Float64,1})
    psi, J = vec0[1], vec0[2]
    cc = (e1-e2)*J
    nk = K - J
    Hchoc = Ec * ( J*(J-1) + nk*(nk-1) )/2
    HTun = -Ej * sqrt(J*nk) * cos(psi)
    H = cc + Hchoc + HTun
    return H
end
```

```julia
function intBEC(tmax::Float64,ai,bi)
    vec0 = [ai, bi]
    time = 0.0
    tV = [time]
    psiV = [vec0[1]]
    JV = [vec0[2]]
    H0 = HBEC(vec0)
    HV = [H0]
    while time < tmax
        h, vec0 = taylorInt(time, vec0, eqsBEC)
        time += h
        push!(tV,time)
        push!(psiV,vec0[1])
        push!(JV,vec0[2])
        H = HBEC(vec0)
        #@show(time, ee)
        push!(HV, H)
        err = abs(H-H0)#/ee0
        #@show time, err
    end
    return tV, psiV, JV, HV
end
```

## Condensado B-E de 3 pozos

### 3 pozos en linea: 

$$ \hat{H} = \epsilon_{1} \hat{a}_{1}^{\dagger} \hat{a}_{1} +  \epsilon_{2} \hat{a}_{2}^{\dagger} \hat{a}_{2} + \epsilon_{3} \hat{a}_{3}^{\dagger} \hat{a}_{3} +\frac{E_{c}}{2} (\hat{a}_{1}^{\dagger 2} \hat{a}_{1}^{2} 
+ \hat{a}_{2}^{\dagger 2} \hat{a}_{2}^{2} + \hat{a}_{3}^{\dagger 2} \hat{a}_{3}^{2}  ) - E_{j} ( \hat{a}_{1}^{\dagger} \hat{a}_{2} + \hat{a}_{2}^{\dagger} \hat{a}_{1} + \hat{a}_{2}^{\dagger} \hat{a}_{3} + \hat{a}_{3}^{\dagger} \hat{a}_{2}) $$

Con:   $$\hat{n}_{i} = \hat{a}_{i}^{\dagger} \hat{a}_{i} $$
En el limite semi-clásico: $$\hat{a}_{i} \rightarrow I_{i}^{\frac{1}{2}} e^{i\phi_{j}} $$  
$$ \hat{a}_{i}^{\dagger} \rightarrow  I_{i}^{\frac{1}{2}} e^{-i\phi_{j}} $$

$$\Longrightarrow \hat{n}_{i} = I_{i} $$


$$ \hat{H} = \epsilon_{1} \hat{n}_{1} +  \epsilon_{2} \hat{n}_{2} + \epsilon_{3} \hat{n}_{3} + \frac{E_{c}}{2} (I_{1}(I_{1}-1) +  I_{2}(I_{2}-1) + I_{3}(I_{3}-1) )
- E_{j} ( I_{1}^{\frac{1}{2}}I_{2}^{\frac{1}{2}}Cos(\phi_{1}-\phi_{2}) + I_{2}^{\frac{1}{2}}I_{3}^{\frac{1}{2}}Cos(\phi_{2}-\phi_{3})) $$


Siendo: $\phi_{i}$ fase de la función de onda en el pozo i, $E_{j}$ la energia asociada al tunneling, $E_{c}$ la energia asociada a los choques entre partículas, $\epsilon_{1}$ y $\epsilon_{2}$ valores asociados con la altura de los pozos. 

### 3 pozos en anillo:

$$ \hat{H} = \epsilon_{1} \hat{a}_{1}^{\dagger} \hat{a}_{1} +  \epsilon_{2} \hat{a}_{2}^{\dagger} \hat{a}_{2} + \epsilon_{3} \hat{a}_{3}^{\dagger} \hat{a}_{3} +\frac{E_{c}}{2} (\hat{a}_{1}^{\dagger 2} \hat{a}_{1}^{2} 
+ \hat{a}_{2}^{\dagger 2} \hat{a}_{2}^{2} + \hat{a}_{3}^{\dagger 2} \hat{a}_{3}^{2}  ) - E_{j} ( \hat{a}_{1}^{\dagger} \hat{a}_{2} + \hat{a}_{2}^{\dagger} \hat{a}_{1} + \hat{a}_{2}^{\dagger} \hat{a}_{3} + \hat{a}_{3}^{\dagger} \hat{a}_{2} +  \hat{a}_{3}^{\dagger} \hat{a}_{1} + \hat{a}_{1}^{\dagger} \hat{a}_{3}) $$

Con:   $$\hat{n}_{i} = \hat{a}_{i}^{\dagger} \hat{a}_{i} $$
En el limite semi-clásico: $$\hat{a}_{i} \rightarrow I_{i}^{\frac{1}{2}} e^{i\phi_{j}} $$  
$$ \hat{a}_{i}^{\dagger} \rightarrow  I_{i}^{\frac{1}{2}} e^{-i\phi_{j}} $$

$$\Longrightarrow \hat{n}_{i} = I_{i} $$

Siendo: $\phi_{i}$ fase de la función de onda en el pozo i, $E_{j}$ la energia asociada al tunneling, $E_{c}$ la energia asociada a los choques entre partículas, $\epsilon_{1}$ y $\epsilon_{2}$ valores asociados con la altura de los pozos.
